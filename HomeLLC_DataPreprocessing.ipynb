```python
monthly_index = pd.date_range(start=df9['date'].min(), end=df9['date'].max(), freq='MS')
monthly_index.shape
```

###Already Monthly Data

```python
df1.set_index('date', inplace=True)
df1 = df1.loc[monthly_index]
df1.shape

df5.set_index('date', inplace=True)
df5 = df5.loc[monthly_index]

df6.set_index('date', inplace=True)
df6 = df6.loc[monthly_index]

df8.set_index('date', inplace=True)
df8 = df8.loc[monthly_index]

df11.set_index('date', inplace=True)
df11 = df11.loc[monthly_index]

df12.set_index('date', inplace=True)
df12 = df12.loc[monthly_index]
```

###Weekly Data (Avg_Mortgage)

```python
df7.head(5)
```

```python
# Convert the 'date' column to datetime objects and set it as the index
df7 = df7.copy() # Make an explicit copy to avoid SettingWithCopyWarning
df7['date'] = pd.to_datetime(df7['date'])
df7.set_index('date', inplace=True)

# Resample to monthly, taking the mean
df7 = df7.resample('MS').mean()

# Align with monthly_index using reindex to handle missing dates
df7 = df7.reindex(monthly_index)
```

###Quaterly data (GDP_per_capita)

```python
#Interpolate to monthly.
df2['date'] = pd.to_datetime(df2['date'])
df2.set_index('date', inplace=True)
df2 = df2.resample('MS').asfreq()  # Create monthly frequency with NaNs
df2['GDP_per_Capita'] = df2['GDP_per_Capita'].interpolate(method='linear')
df2 = df2.reindex(monthly_index)
```

```python
df10['date'] = pd.to_datetime(df10['date'])
df10.set_index('date', inplace=True)
df10 = df10.resample('MS').asfreq()
df10['total_households'] = df10['total_households'].interpolate(method='linear')
df10 = df10.reindex(monthly_index)

df4['date'] = pd.to_datetime(df4['date'])
df4.set_index('date', inplace=True)
df4 = df4.resample('MS').asfreq()
df4['median_household_income'] = df4['median_household_income'].interpolate(method='linear')
df4 = df4.reindex(monthly_index)

df3['date'] = pd.to_datetime(df3['date'])
df3.set_index('date', inplace=True)
df3 = df3.resample('MS').asfreq()
df3['housing_subsidies'] = df3['housing_subsidies'].interpolate(method='linear')
df3 = df3.reindex(monthly_index)
```
